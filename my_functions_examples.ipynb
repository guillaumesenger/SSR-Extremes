````python
# Load example time daily mean SSR time series 
# Specificaly, this is the 500-year piControl run by GFDL-ESM4 model
time_series = pickle.load(open(f'Data/SSR_GFDL_ESM4.pickle', 'rb'))

# Compute SREs
output_SREs = my_functions.compute_SREs(time_series = time_series, percentiles = 'all', min_length = 3, max_length = 16)
````

### Outputs illustration

````python
percentile = 85 # Choose the percentile 

x = np.arange(365)

fig = plt.figure(figsize = (20,25))
gs  = fig.add_gridspec(nrows=5, ncols=2, width_ratios = [2,0.28], height_ratios = [1,1,1,0.2,1])

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[2,0])
ax4 = fig.add_subplot(gs[4,:])

ax = ax1 # Plot the time series and the threshold for 1 year

ax.set_title('year 1/500 SREs analysis outputs', fontsize = 22)
ax.plot(x, time_series[0:365], label = 'time series')
ax.plot(x, output_SREs[percentile]['thresholds_all_years'][0:365], label = 'daily threshold')
ax.legend()
ax.set_ylabel('daily mean SSR [W/m$^2$]')

ax = ax2 # Plot the rsds_all_years_above time series for 1 year

ax.plot(x, output_SREs[percentile]['rsds_all_years_above'][0:365], label = '1 = above threshold, 0 = not above threshold')
ax.legend()


ax = ax3 # Plot the SREs_all_years time series for 1 year

ax.plot(x, output_SREs[percentile]['SREs_all_years'][0:365], label = 'SREs lengths (only if 3 or more consecutive days)')
ax.legend()
ax.set_xlabel('days')
ax.set_ylabel('SREs length [days]')


ax = ax4 # Plot the extreme_events_per_day array for 1 year

ax.set_title('Number of SREs of different length starting on \n each calendar day for the entire time series', fontsize = 22)
to_plot = output_SREs[percentile]['extreme_events_per_day']

to_plot[to_plot == 0] = np.nan
plot = ax.pcolormesh(to_plot.T, cmap = 'jet', vmin = 1)
colorbar = plt.colorbar(plot, label = 'number of SREs')

ax.set_xlabel('calendar days')
ax.set_ylabel('SREs length [days]')

for ax in [ax1, ax2, ax3, ax4]:
    ax.grid()
    ax.set_xlim(0,364)

plt.show()
````
